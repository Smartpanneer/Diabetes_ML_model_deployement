```python
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
df=pd.read_csv('diabetes.csv')
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.size
```

```python
df.shape
```

```python
x=df.drop('Outcome',axis=1)
x
```

```python
y=df.Outcome
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

# Feature Scaling - Important for Logistic Regression
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.size
```

```python
y_test.size
```

```python
# Model with improved hyperparameters
model = LogisticRegression(max_iter=1000, random_state=42, C=0.1)
```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
ev = accuracy_score(y_test, pred)
precision = precision_score(y_test, pred)
recall = recall_score(y_test, pred)
f1 = f1_score(y_test, pred)

print(f"Accuracy: {ev:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, pred))
print("\nClassification Report:")
print(classification_report(y_test, pred))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC

# First, use unscaled data for tree-based models
x_train_unscaled, x_test_unscaled, y_train_unscaled, y_test_unscaled = train_test_split(x, y, test_size=0.2, random_state=42)

# Random Forest Model
print("=" * 50)
print("Random Forest Classifier")
print("=" * 50)
rf = RandomForestClassifier(n_estimators=200, max_depth=15, min_samples_split=2, random_state=42)
rf.fit(x_train_unscaled, y_train_unscaled)
pred_rf = rf.predict(x_test_unscaled)

print(f"Accuracy: {accuracy_score(y_test_unscaled, pred_rf):.4f}")
print(f"Precision: {precision_score(y_test_unscaled, pred_rf):.4f}")
print(f"Recall: {recall_score(y_test_unscaled, pred_rf):.4f}")
print(f"F1-Score: {f1_score(y_test_unscaled, pred_rf):.4f}")

# Gradient Boosting Model
print("\n" + "=" * 50)
print("Gradient Boosting Classifier")
print("=" * 50)
gb = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)
gb.fit(x_train_unscaled, y_train_unscaled)
pred_gb = gb.predict(x_test_unscaled)

print(f"Accuracy: {accuracy_score(y_test_unscaled, pred_gb):.4f}")
print(f"Precision: {precision_score(y_test_unscaled, pred_gb):.4f}")
print(f"Recall: {recall_score(y_test_unscaled, pred_gb):.4f}")
print(f"F1-Score: {f1_score(y_test_unscaled, pred_gb):.4f}")

# SVM with scaling
print("\n" + "=" * 50)
print("Support Vector Machine (SVM)")
print("=" * 50)
svm = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
svm.fit(x_train, y_train)  # Using scaled data
pred_svm = svm.predict(x_test)

print(f"Accuracy: {accuracy_score(y_test, pred_svm):.4f}")
print(f"Precision: {precision_score(y_test, pred_svm):.4f}")
print(f"Recall: {recall_score(y_test, pred_svm):.4f}")
print(f"F1-Score: {f1_score(y_test, pred_svm):.4f}")
```

```python
import pickle

# Save Logistic Regression Model
lr_model = LogisticRegression(max_iter=1000, random_state=42, C=0.1)
lr_model.fit(x_train, y_train)

# Save model to pickle file
with open('logistic_regression_model.pkl', 'wb') as file:
    pickle.dump(lr_model, file)

# Save scaler to pickle file
with open('scaler.pkl', 'wb') as file:
    pickle.dump(scaler, file)

print("✓ logistic_regression_model.pkl saved successfully!")
print("✓ scaler.pkl saved successfully!")

# Test loading to verify
with open('logistic_regression_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
    
with open('scaler.pkl', 'rb') as file:
    loaded_scaler = pickle.load(file)

# Verify model works
test_pred = loaded_model.predict(x_test)
test_accuracy = accuracy_score(y_test, test_pred)
print(f"\n✓ Model loaded successfully! Test Accuracy: {test_accuracy:.4f}")
```
